```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import galpy as gal
import astropy.coordinates as coord
from astropy import units as u
from astropy import io
from scipy import stats
import matplotlib as mplt
```

```python
plt.style.use(["default"])
plt.rcParams['figure.figsize'] = 8.0,8.0
font = {'family' : 'Sans',
        'weight' : 'normal',
        'size'   :18}
plt.rc("font",**font)
```

```python
completo = pd.DataFrame(pd.read_csv("galah_complete.csv"))
```

```python
completo = completo.loc[(completo["teff"].notna())&(completo["logg"].notna())]
```

```python
mplt.cm.get_cmap("viridis")(0)
```

```python
h = plt.hist2d(completo["teff"], completo["logg"], bins=500, cmin=5, norm=mplt.colors.PowerNorm(0.5), zorder=4,cmap=plt.cm.viridis)
plt.scatter(completo["teff"],completo["logg"],s=1,color=mplt.cm.get_cmap("inferno")(0),zorder=0)
#cb = plt.colorbar(h[3], pad=0.02)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.ylabel("$\log(g)$")
plt.xlabel("T$_{eff}$ [K]")
#plt.ylim(5,3)
plt.show()
```

```python
giants = completo.loc[(completo["teff"] < 5500) & (completo["logg"] < 3)]
```

```python
h = plt.hist2d(giants["teff"], giants["logg"], bins=300, cmin=5, norm=mplt.colors.PowerNorm(0.5), zorder=4)
plt.scatter(giants["teff"],giants["logg"],s=1,color=mplt.cm.get_cmap("inferno")(0),zorder=0)
#cb = plt.colorbar(h[3], pad=0.02)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.ylabel("$\log(g)$")
plt.xlabel("T$_{eff}$ [K]")
#plt.ylim(5,3)
plt.show()
```
