```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib as mplt
import seaborn as sns
```

```python
plt.style.use("default")
plt.rcParams['figure.figsize'] = 8.0,8.0
```

```python
enceladus = pd.DataFrame(pd.read_csv("enceladus.csv"))
```

```python
completo = pd.DataFrame(pd.read_csv("complete.csv"))
```

```python
vecindad = pd.DataFrame(pd.read_csv("solarnei.csv"))
```

```python
completo = completo.loc[(completo["phot_bp_mean_mag"].notna()) & (completo["phot_rp_mean_mag"].notna())]
vecindad = vecindad.loc[(vecindad["phot_bp_mean_mag"].notna()) & (vecindad["phot_rp_mean_mag"].notna())]
```

```python
G = enceladus["phot_g_mean_mag"] + 5 +5*np.log10(enceladus["parallax"]/1000)
G_comp = completo["phot_g_mean_mag"] + 5 + 5*np.log10(completo["parallax"]/1000)
G_vec = vecindad["phot_g_mean_mag"] + 5 +5*np.log10(vecindad["parallax"]/1000)
```

```python
h = plt.hist2d(completo["phot_bp_mean_mag"]-completo["phot_rp_mean_mag"], G_comp, bins=500, cmin=5, 
               norm=mplt.colors.PowerNorm(0.5), zorder=0,cmap=plt.cm.inferno)
plt.scatter(enceladus["phot_bp_mean_mag"]-enceladus["phot_rp_mean_mag"],G,color="red",zorder=4,s=5)
plt.scatter(completo["phot_bp_mean_mag"]-completo["phot_rp_mean_mag"],G_comp,s=1,color="black",alpha=0.4,zorder=-1)
plt.gca().invert_yaxis()
plt.xlabel("$\mathrm{G_{BP}-G_{RP}}$",fontsize=14)
plt.ylabel("G",fontsize=14)

plt.show()
```

```python
sub_vecindad = vecindad.loc[(vecindad["fe_h"].notna())&(vecindad["Mg_fe"].notna())]
```
